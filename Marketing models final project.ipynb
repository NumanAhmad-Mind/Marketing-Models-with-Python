```python
#T-test among every variable

#array = np.array((len(data), len(data)))

for i in data.columns:
    for j in data.columns:
        t, p = ttest_ind(data[i], data[j])
        #array[data.index(i)][data.index(j)] = p
        if p < 0.05:
            print(i, "and",j ,"have a significant difference")

           # print("p-value:",p)
```

```python
for i in data.columns[9:]:
    for j in data.columns[8:]:
        if i != j:
            t, p = ttest_ind(data[i], data[j],equal_var = False)

            if p < 0.05:
                print(i, "and",j ,"have a significant difference")

            else:
                print(i, "and",j ,"do not have a significant difference")
            
            print("p-value:",p)
            print()
```

```python
i = "career-mentorship"
j = "office hours"

t, p = ttest_ind(data[i], data[j],equal_var = False)   

if p < 0.05:
    print(i, "and",j ,"have a significant difference")

else:
    print(i, "and",j ,"do not have a significant difference")
print("p-value:",p)
```
